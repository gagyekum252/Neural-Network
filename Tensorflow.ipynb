```python
#Shuffle a tensor (valuable for when you want to shuffle your data so the inherent order doesn't affect order

not_shuffled = tf.constant([[10,7],
                           [3,4],
                           [2,5]])
#shuffle our non-shuffled tensor
not_shuffled.ndim
tf.random.shuffle(not_shuffled)
```

```python
#shuffle our non-shuffled tensor
tf.random.set_seed(42)
tf.random.shuffle(not_shuffled, seed =42)
```

**Global Seed: The tf.random.set_seed(seed) function sets a global seed for all random operations. This ensures that the sequence of random numbers is deterministic across program runs.**

**Operation-Level Seed: If you also specify a seed at the operation level (e.g., tf.random.uniform(shape, seed=123)), both the global and operation seeds are combined to generate a unique sequence.**

**Reproducibility: When only the global seed is set, TensorFlow ensures that the same sequence of random numbers is generated for each program run.**

**Important Considerations**

**If neither a global nor an operation-level seed is set, TensorFlow generates different results for every run.**

**For complete reproducibility across different environments or versions, explicitly set both global and operation-level seeds.**

**Use tf.random.stateless_uniform for stateless random operations if you need deterministic behavior without relying on internal counters.**

**This approach is particularly useful in machine learning workflows where reproducibility is critical for debugging and experimentation.**

### Getting information from our tensors
When dealing with tensors you probably want to be aware of the following attributes:

* Shape - the length(number of elements) of each of the dimensions of a tensor
* Rank - the number of tensor dimensions. A scalar has rank 0 a vector has rank 1, a matrix is rank 2, a tensor has rank n
* Axis or dimension - A particular dimension of a tensor.  tensor[0], tensor[:, 1]
* Size - total number of items in the tensor  tf.size()

```python
# Create a Rank 4 tensor (4 dimensions)
rank_4_tensor = tf.zeros([2, 3, 4, 5])
rank_4_tensor
```

```python
rank_4_tensor[0]
```

```python
rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor)
```

```python
# Get various attributes of our tensor
print('Datatype of every element:', rank_4_tensor.dtype)
print('Number of dimensions (rank):', rank_4_tensor.ndim)
print('Shape of the tensor:', rank_4_tensor.shape)
print('Elements along the 0 axis:', rank_4_tensor.shape[0])
print('Elements along the last axis:', rank_4_tensor.shape[-1])
print('Total number of elements in our tensor:', tf.size(rank_4_tensor))
print('Total number of elements in our tensor:', tf.size(rank_4_tensor).numpy())
```

**Matrix Multiplication**

In machine learing, matrix multiplication is one of the most common tensor operations

http://matrixmultiplication.xyz/

Resource: Info and example of matrix multiplication: https://www.mathsisfun.com/algebra/matrix-multiplying.html

There are two rules our tensors (or matrices ) need to fulfil if we are going to matrix multiply them:
1. The inner dimension must match
2. The resulting matrix has the shape of the outer dimensions

```python
# Matrix multiplication in tensorflow
print(tensor)
tf.matmul(tensor, tensor)
```

```python
# Matrix multiplication with python operator "@". the @ operator is tensor oprator in python
tensor @ tensor
```

```python
# Create a tensor (3,2) tensor
x = tf.constant([[1,2],
                [3,4],
                [5,6]])

#Create another (3,2) tensor
y = tf.constant([[7,8],
                [9,10],
                [11,12]])
x,y
```

```python
# Try to matrix multiply tensors of same shape
x @ y
tf.matmul(x,y)  # This will produce an error because the inner dimension must match
```

```python
# Let change the shape of Y
tf.reshape(y, shape = (2,3))
```

```python
x.shape, tf.reshape(y, shape=(2,3)).shape
```

```python
# Try to multiply X by reshape y
x @ tf.reshape(y, shape=(2,3))
```

```python
tf.matmul(x, tf.reshape(y, shape=(2,3)))
```

```python
tf.reshape(x, shape=(2,3)).shape, y.shape
```

```python
# Try change the shape of x instead of y
tf.matmul(tf.reshape(x, shape=(2,3)),y)
```

```python
# Can do the same with transpose
x, tf.transpose(x), tf.reshape(x, shape=(2,3))
```

```python
# Try matrix multiplication with transpose rather than reshape
tf.matmul(tf.transpose(x), y)
```

**Find the variance and standard deviation of our E tensor using tensorFlow methods**

```python
 import tensorflow_probability as tfp
```

```python
# Find the variance of our tensor we need access to tensorflow_probability

tfp.stats.variance(E)
```
